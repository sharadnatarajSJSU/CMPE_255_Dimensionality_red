```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("/content/breast_cancer.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
sc = StandardScaler()
X_scaled = sc.fit_transform(df)
```

```python
pca = PCA(n_components=6)
X_PCA=pca.fit_transform(X_scaled)
```

```python
print("EigenValues")
print(pca.explained_variance_)
```

```python
print(pca.explained_variance_ratio_)
```

```python
print(pca.components_)
```

```python
dataframe = pd.DataFrame({'variance':pca.explained_variance_ratio_,
             'Principal Component':['PC1','PC2','PC3','PC4','PC5','PC6']})
sns.barplot(x='Principal Component',y="variance", 
           data=dataframe, color="b");
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of principal components')
plt.ylabel('Cumulative explained variance')
plt.show()
```

#SVD

```python
u, s, v = np.linalg.svd(X_scaled, full_matrices=True)
```

```python
u.shape
```

```python
s.shape
```

```python
v.shape
```

```python
variance_explained = np.round(s**2/np.sum(s**2), decimals=3)
variance_explained
 
sns.barplot(x=list(range(1,len(variance_explained)+1)),
            y=variance_explained, color="blue")
plt.xlabel('SVs', fontsize=16)
plt.ylabel('Percent of the variance explained', fontsize=15)
```

#LLE

```python
from sklearn.manifold import LocallyLinearEmbedding
model = LocallyLinearEmbedding(n_neighbors=100, n_components=2, method='modified',
                               eigen_solver='dense')
out = model.fit_transform(X_scaled)

fig, ax = plt.subplots()
ax.scatter(out[:, 0], out[:, 1])
ax.set_ylim(.5
, -0.15);
```
